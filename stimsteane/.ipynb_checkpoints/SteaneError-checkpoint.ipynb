## 2.2 Simulation pipeline

The simulation pipeline consists of two steps:
- Definition of circuit
- Calculation of logical error rate

```python
def decode_to_error_rate(circuit: stim.Circuit, num_shots: int) -> float:
    """
    Calculate logical error rate given Stim Circuit and number of shots.

    The Circuit must contain Detectors and at least one Observable, in order to
    detect errors.
    """
    
    sampler = circuit.compile_detector_sampler()
    detection_events, observable_flips = sampler.sample(
        num_shots, separate_observables=True
    )
    
    dem = circuit.detector_error_model()
    
    matcher = pymatching.Matching.from_detector_error_model(dem)
    predictions = matcher.decode_batch(detection_events)
    
    num_errors = 0
    for shot in range(num_shots):
        actual_for_shot = observable_flips[shot]
        predicted_for_shot = predictions[shot]
        if not np.array_equal(actual_for_shot, predicted_for_shot):
            num_errors += 1
    return num_errors / num_shots
```

This will not produce any error, since no error source is introduced.

```python
print(decode_to_error_rate(steane_circuit, 100_000))
```

Let's introduce errors again. This is done by utilizing a Noise module, that modifies the specified circuit. Depolarization on 1- and 2 qubit gates, as well as measurement errors are introduced.

The introduced errors both affect the measurement circuit and initial encoding circuit. Additionally, measurement errors are present. The latter error directly leads to a incorrect syndrom measurement.

## 2.3 Determining the Error-Threshold
In the following section, the logical-error rate as a function of the physical-error rate is estimated via sampling.

In the utilized noise model it is possible to defined both depolarization on 1- and 2-qubit gates. Let's sample over different noise amplitudes for these errors, and plot the resuling error-rates.

### 2.3.1 Error Threshold

As the error threshold generally refers to the scaling behavior as the code distance increases, another color code needs to be introduced. For this, the pre-defined *color_code:memory_xyz* is selected. In both circuits, only 1-qubit depolarization error is present.

When running the experiment, a diverging behaviour between the two color codes is expected, when reaching the error threshold.


```python
from codes.steane_713 import get_steane_713_noisy_code

def get_circuit(d, noise):
    if d == 3:
        circuit = get_steane_713_noisy_code(meas_error = 0, depolarize2 = 0, depolarize1 = noise)
    elif d == 5:
        circuit = circuit = stim.Circuit.generated(
            "color_code:memory_xyz",
            rounds=2*d,
            distance=d,
            before_round_data_depolarization=noise,
        )
    else:
        pass

    return circuit
    
# Define and run tasks
steane_code_tasks = [
    sinter.Task(
        circuit = get_circuit(distance, noise),
        json_metadata={'noise': noise, 'distance': distance},
    )
    for distance in [3, 5]
    for noise in [1e-1, 5e-2, 1e-2, 5e-3, 1e-3, 1e-4]
]

collected_steane_code_stats: List[sinter.TaskStats] = sinter.collect(
    num_workers=os.cpu_count(),
    tasks=steane_code_tasks,
    decoders=['pymatching'],
    max_shots=1_000_000,
    max_errors=50_000,
    print_progress=False,
)

# Plot results
fig, ax = plt.subplots(1, 1)
sinter.plot_error_rate(
    ax=ax,
    stats=collected_steane_code_stats,
    x_func=lambda stat: stat.json_metadata['noise'],
    group_func=lambda stat: stat.json_metadata['distance']
)

ax.axvline(x=3*1e-3, color='black', linestyle='--')
plt.gca().invert_xaxis()
ax.loglog()
ax.set_xlabel("Phyical Error Rate Scaling")
ax.set_ylabel("Logical Error Rate")
ax.grid(which='minor')
ax.legend(title="Distance")
fig.set_dpi(120)
```

It is now possible to estimate the error threshold to be around 3e-3 ~= 0.003%, which is very low.  

Nevertheless, since the Steane code has a fixed distance 𝑑 = 3, it does not have a threshold in the strict sense, as it is not directly possible to increase the distance. In the following section, a pseudo-threshold is introduced. Following this, the necessary error level is determined, in order to satisfy the pre-defined pseudo-threshold.

### 2.3.2 Pseudo-Threshold

For QEC to be helpful in computations, logical level error rates must be below the physical level error rates.
In cases where the dominant physical noise mechanism is known (the p2_error), one method is to scale that noise mechanism in simulations and solve for the pseudothreshold, defining it as the point where the logical error rate is equal to the dominant physical noise source.

This is a reproduction of experiments in "Realization of Real-Time Fault-Tolerant Quantum Error Correction". The Pseudo-Threshold is set to the physical error-scaling * p2, with p2_error being the 2-qubit gate depolarization error.

This experiment is a coarse view of how much the dominating physical error (the two-qubit gate) needs to improve to reach the simulated pseudothreshold. During simulation, all physical error rates are scaled by a common scaling factor. 

```python
measurement_error = 0 # Assume Perfect Read-Out (Remark: For values ~1e-3, influence is very small and hardly noticable)

d2s = 1
d1s = 1
# Define and run tasks
steane_code_tasks = [
    sinter.Task(
        circuit = get_steane_713_noisy_code(meas_error = 0, depolarize2 = 3.1 * 1e-3, depolarize1 = 1e-5,
                                            depolarize2_scaling = d2s, depolarize1_scaling = d1s, scaling_general=gs),
        json_metadata={'gs': gs, 'd2s': d2s},
    )
    for d2s in [1, 1e-1]
    for gs in [1.5, 1, 0.75, 0.5, 0.25, 1e-1, 1e-2, 1e-3, 1e-4]
]

collected_steane_code_stats: List[sinter.TaskStats] = sinter.collect(
    num_workers=os.cpu_count(),
    tasks=steane_code_tasks,
    decoders=['pymatching'],
    max_shots=10_000_000,
    max_errors=500_000,
    print_progress=False,
)
```

```python
# Plot results
fig, ax = plt.subplots(1, 1)
sinter.plot_error_rate(
    ax=ax,
    stats=collected_steane_code_stats,
    x_func=lambda stat: stat.json_metadata['gs'],
    group_func=lambda stat: stat.json_metadata['d2s']
)

# Plot pseudothreshold
x = np.linspace(1e-4,1, 500)
y = x * 3.1e-3
plt.plot(x, y, 'k--', label='Pseudo-Threshold')

ax.axvline(x=5*1e-4, color='black', linestyle=':')

plt.gca().invert_xaxis()
ax.loglog()
ax.set_xlim(1, 1e-4)
ax.set_title("Comparison of physical and logical error rates")
ax.set_xlabel("Phyical Error Rate Scaling")
ax.set_ylabel("Logical Error Rate")
#ax.grid(which='major')
ax.grid(which='minor')
ax.legend(['1/10 p2 Depolarization Error', '1-STD', 'Unscaled Error Parameters', '1-STD', 'p2', 'Threshold Satisfied'], title="Dephasing rate")
fig.set_dpi(120)
```

Different to the aforementioned paper, the reduction of p2_error is much higher than presented.
For the utilized noise model in the experiments above, the pseudo-threshold is reached at around 5*10^-3.
This likely stems from the different noise models utilized.
